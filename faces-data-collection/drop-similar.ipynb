```python
import os
from PIL import Image
import numpy as np
import torch
import faiss
from transformers import AutoImageProcessor, AutoModel, AutoProcessor
from facenet_pytorch import MTCNN, InceptionResnetV1
from torchvision.transforms.functional import pil_to_tensor
import json
from tqdm.auto import tqdm


model_id = 'facebook/dinov2-base'

device = torch.device('cuda' if torch.cuda.is_available() else "cpu")
processor = AutoImageProcessor.from_pretrained(model_id)
model = AutoModel.from_pretrained(model_id).to('cuda:0')

mtcnn = MTCNN()
resnet = InceptionResnetV1(pretrained='vggface2').eval() #авторы - уебки, будет долгий инференс на cpu
```

```python
image_files = []

for i in range(4):
    with open(f'output_aigen_part_{i}.json', 'r') as f: #ffhq valid
        image_files.extend(json.loads(f.read())['paths'])
    
image_files = list(set(image_files))
len(image_files)
```

```python
def get_dino_emb(image_files):
    vectors = []
    valid_paths = []
    for filename in tqdm(image_files):
        try:
            image = Image.open(filename).convert('RGB')
            inputs = processor(images=image, return_tensors="pt").to(device)
            with torch.no_grad():
                outputs = model(**inputs)
            embeddings = outputs.last_hidden_state.mean(dim=1)
            vector = embeddings.cpu().numpy().astype(np.float32)
            faiss.normalize_L2(vector)
            vectors.append(vector[0])
            valid_paths.append(filename)
        except:
            pass
    return np.vstack(vectors), valid_paths

def gen_mtcnn_emb(image_files): 
    vectors = []
    valid_paths = []
    
    for filename in tqdm(image_files):
        try:
            image = Image.open(filename).convert('RGB')
            embeddings = resnet(mtcnn(image).unsqueeze(0)).detach().numpy()
            vector = embeddings.astype(np.float32)
            faiss.normalize_L2(vector)
            vectors.append(vector[0])
            valid_paths.append(filename)
        except:
            pass
    return np.vstack(vectors), valid_paths
    

def show_results(image_files, similar_images, N=10):
    c = 0
    for i in similar_images:
        if c == N:
            break
        display(Image.open(image_files[i[0]] ).resize((64, 64)))
        display(Image.open(image_files[i[1]]).resize((64, 64)))
        c += 1

def clean_data(image_files, vectors, index, n_neighbors = 1000, batch_size = 1000, threshold = 0.45):
    similar_images = []

    for i in tqdm(range(0, len(vectors), batch_size)):
        batch_data = vectors[i:i + batch_size]
        _, indices = index.search(batch_data, n_neighbors)

        for j in range(batch_data.shape[0]):
            similar_indices = indices[j]

            for sim_index in similar_indices:
                if sim_index == i + j:continue
                distance = np.sqrt(2 - 2 * np.clip(np.dot(vectors[i + j], vectors[sim_index]), -1, 1))
                if distance < (1 - threshold):
                    similar_images.append((i + j, sim_index))
                else:
                    break

    indices_to_remove = set()

    for idx1, idx2 in similar_images:
        idx1 = image_files[idx1]
        idx2 = image_files[idx2]
        if idx1 not in indices_to_remove:
            indices_to_remove.add(idx2)

    return similar_images, [i for i in image_files if i not in indices_to_remove]
```

```python
index = faiss.IndexFlatL2(768)
vectors, file_paths = get_dino_emb(image_files)
index.add(vectors)
```

```python
similar, cleaned_paths = clean_data(file_paths, vectors, index, n_neighbors = 1000, batch_size = 1000, threshold = 0.6)


show_results(file_paths, similar)

print(len(file_paths), len(cleaned_paths))
```

```python
index_2 = faiss.IndexFlatL2(512)
vectors_2, file_paths_2 = gen_mtcnn_emb(cleaned_paths)
index_2.add(vectors_2)
```

```python
similar_2, cleaned_paths_2 = clean_data(file_paths_2, vectors_2, index_2, n_neighbors = 1000, batch_size = 1000, threshold = 0.4)


show_results(file_paths_2, similar_2)
print(len(cleaned_paths_2))
```
